## Extracting metadata_config from Metadata Table

```python
import json

metadata_df = spark.table("ottconsumption.bronze.metadata_config") \
                   .filter("is_active = 'Y'")

rows = [row.asDict() for row in metadata_df.collect()]

dbutils.jobs.taskValues.set(
    key="metadata_list",
    value=json.dumps(rows)
)
```
